```python
from Bio import SeqIO

input_file = r"D:\Project\Drug Target\Proteom\Proteome.fasta.faa"
output_file = r"D:\Project\Drug Target\Proteom\proteome_NR.fasta"

unique_sequences = {}
unique_records = []

for record in SeqIO.parse(input_file, "fasta"):
    seq = str(record.seq)

    if seq not in unique_sequences:
        unique_sequences[seq] = record.id
        unique_records.append(record)

SeqIO.write(unique_records, output_file, "fasta")

print("Original proteins:", len(list(SeqIO.parse(input_file, "fasta"))))
print("Non-redundant proteins:", len(unique_records))
```
